```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('RR.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

 CALCULATE THE ATTRITION RATE

```python
total_employees = len(df)
employees_left = df[df['Attrition'] == 'Yes'].shape[0]
attrition_rate = (employees_left / total_employees) * 100
```

```python
print(f"Total Employees: {total_employees}")
print(f"Employees Who Left: {employees_left}")
print(f"Attrition Rate: {attrition_rate:.2f}%\n")
```

ANALYZE FACTORS: AGE, YEARS, INCOME

```python
summary = df.groupby('Attrition')[['Age', 'YearsAtCompany', 'MonthlyIncome']].mean().round(2)
print("Average Comparison by Attrition Status:")
print(summary, "\n")
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(5,4))
sns.countplot(data=df, x='Attrition', hue='Attrition', palette='Set2', legend=False)
plt.title("Overall Attrition Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Attrition', y='Age', hue='Attrition', palette='coolwarm', legend=False)
plt.title("Age vs Attrition")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Attrition', y='YearsAtCompany', hue='Attrition', palette='viridis', legend=False)
plt.title("Years at Company vs Attrition")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome', hue='Attrition', palette='magma', legend=False)
plt.title("Monthly Income vs Attrition")
plt.show()
```

## Key Insights:
### Average age of employees who left is generally lower.
### Employees with fewer years at the company show higher attrition.
### Lower-income employees are more likely to leave.
